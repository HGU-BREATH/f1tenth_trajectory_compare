```python
import matplotlib.pyplot as plt
import numpy as np
import sys

def main(file1, file2):
    
    data1 = np.genfromtxt(file1, delimiter=',', skip_header=1)  #   skip_header=1 
    if(data1.shape[1] == 2):
        x1 = data1[:, 0]  # x    
        y1 = data1[:, 1]  # y    
    else:
        x1 = data1[:, 1] 
        y1 = data1[:, 2] 

    data2 = np.genfromtxt(file2, delimiter=',', skip_header=1)
    if(data2.shape[1] == 2):
        x2 = data2[:, 0]  # x    
        y2 = data2[:, 1]  # y    
    else:
        x2 = data2[:, 1] 
        y2 = data2[:, 2] 
    
    
    #  
    plt.figure(figsize=(20, 15))  #   
    
    plt.plot(x1, y1, marker='o', linewidth=5, color='red', alpha=1, label='Accumulated Trajectory')  #     
    plt.plot(x2, y2, marker='o', linewidth=2, color='blue', alpha=0.5, label='Average Trajectory')
    
    plt.title('Trajectory Visualization')  #  
    plt.xlabel('X Position')  # x 
    plt.ylabel('Y Position')  # y 
    plt.grid(True)  #  
    plt.show()  #  

if __name__ == "__main__":

    file1 = input("Enther the accumulated trajectory CSV file path: ")
    file2 = input("Enter the average trajectory CSV file path: ")
    
    main(file1, file2)
```
